```python
alphabets = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
len(alphabets)
```

```python
int2char = {idx:char for idx,char in enumerate(alphabets)}
int2char
```

```python
char2int = {char:idx for idx,char in int2char.items()}
char2int
```

```python
seq_length = 1
x = []
y = []
for i in range(len(alphabets)-seq_length):
    seq_in = [alphabets[i]]
    seq_out = alphabets[i+1]
    x += [[char2int[seq_in[0]]]]
    y += [char2int[seq_out]]
    print(seq_in,'->',seq_out)
```

```python
from keras.utils import to_categorical
```

```python
y_new = to_categorical(y)
```

```python
### Build the model
from keras.models import Sequential
from keras.layers import Dense,LSTM
```

```python
model = Sequential()
model.add(Dense(25,input_shape=(1,1),activation='relu'))
model.add(LSTM(128))
model.add(Dense(26,activation='softmax'))
```

```python
model.compile(loss="categorical_crossentropy",optimizer='adam',metrics=['accuracy'])
```

```python
model.fit(x,y_new,epochs=1000,batch_size=1)
```
