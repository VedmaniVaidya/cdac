```python
def plot_images(dataset ,num_images=25):
    # Get a batch of images and labels from the dataset

    plt.figure(figsize=(10, 10))
   
    for i in range(num_images):
        ax = plt.subplot(5, 5, i + 1)
        plt.imshow(dataset[i],cmap="gray")
        plt. grid(False)
        plt.xticks([])
        plt.yticks([])

plot_images(dataset=X_train)
```
