Interest points:
- A point in an image that has a well-defined position and can be robustly detected in different images.
- Typically associated with a significant change in one or more image properties simultaneously (e.g. intensity, color, texture)

```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
import cv2
```

```python
img = cv2.imread('./images/dog.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
sift = cv2.SIFT_create()
kp, _ = sift.detectAndCompute(gray, None)
new_img = cv2.drawKeypoints(gray, kp, img, flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
plt.imshow(new_img)
```
