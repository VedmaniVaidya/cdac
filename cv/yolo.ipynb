```python
import torch
import cv2
import torchvision
import numpy as np
import matplotlib.pyplot as plt
```

```python
#yolo5 
model = torch.hub.load('ultralytics/yolov5', 'yolov5s', pretrained=True)
```

```python
image = cv2.imread('./data/images/jungle.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
results = model(image)
```

```python
#display the results in form of image
results.show()
```
