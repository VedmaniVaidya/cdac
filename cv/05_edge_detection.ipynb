* Edege detection is a technique of images processing to identify points in a digital image with discontinuities, simplyto say, sharp changes in image brightess.
* These points where the brightness varies sharply are called the edges (or the boundaries) of the image.

```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
import cv2
```

```python
#load draw image
img = cv2.imread('./images/draw.png')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
#apply sobel filter
gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
sobel_x = cv2.Sobel(gray, cv2.CV_64F, 1, 0, ksize=5)
sobel_y = cv2.Sobel(gray, cv2.CV_64F, 0, 1, ksize=5)
sobel_x = np.uint8(np.absolute(sobel_x))
sobel_y = np.uint8(np.absolute(sobel_y))
sobel = cv2.bitwise_or(sobel_x, sobel_y)
plt.imshow(sobel, cmap='gray')
```

```python
#perform canny edge detection
low_threshold = 100
high_threshold = 200
canny = cv2.Canny(gray, low_threshold, high_threshold)
plt.imshow(canny, cmap='gray')
```

```python
#load buildings image and apply canny edge detection
img = cv2.imread('./images/buildings.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
low_threshold = 100
high_threshold = 200
canny = cv2.Canny(gray, low_threshold, high_threshold)
plt.imshow(canny, cmap='gray')
```

```python
#using pillow library
from PIL import Image, ImageFilter
image = Image.open('./images/draw.png')
image = image.convert('L')
plt.imshow(image, cmap='gray')
```
