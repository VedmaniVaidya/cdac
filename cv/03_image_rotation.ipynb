```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
import cv2
```

```python
img = cv2.imread('./images/caves.jpg')
img.shape
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
img.transpose().shape
```

```python
plt.imshow(img)
```

```python
width, height = img.shape[:2]
```

```python
rotation_matrix = cv2.getRotationMatrix2D((width/2, height/2), 180, 1)
rotated_image = cv2.warpAffine(img, rotation_matrix, (width, height))
```

```python
plt.imshow(rotated_image)
```

```python
#rotate image using pillow
img = Image.open('./images/caves.jpg')
rotated_image = img.rotate(180)
plt.imshow(rotated_image)
```

```python
#resize image
img = cv2.imread('./images/caves.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img.shape

new_image = cv2.resize(img, (1200, 600))
plt.imshow(new_image)
```
