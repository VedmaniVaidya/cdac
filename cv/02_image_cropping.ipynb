```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
import cv2
```

```python
img = cv2.imread('./images/dog.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

## Relative cropping

```python
height, width, _ = img.shape
height, width
```

```python
start_row, start_col = int(height * .25), int(width * .25)
end_row, end_col = int(height * .75), int(width * .75)

cropped = img[start_row:end_row, start_col:end_col]

plt.imshow(cropped)
cropped = cv2.cvtColor(cropped, cv2.COLOR_BGR2RGB)
```

```python
#devide image in 4 parts and display them in subplots
fig, axs = plt.subplots(2, 2)
axs[0, 0].imshow(img[0:int(height/2), 0:int(width/2)])
axs[0, 1].imshow(img[0:int(height/2), int(width/2):width])
axs[1, 0].imshow(img[int(height/2):height, 0:int(width/2)])
axs[1, 1].imshow(img[int(height/2):height, int(width/2):width])
```
